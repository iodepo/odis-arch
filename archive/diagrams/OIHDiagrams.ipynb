## Architecture


````python
from diagrams import Cluster, Diagram, Node, Edge
from diagrams.gcp.analytics import BigQuery, Dataflow, PubSub
from diagrams.gcp.compute import AppEngine, Functions
from diagrams.gcp.database import BigTable
from diagrams.programming.language import Go
from diagrams.onprem.container import Docker
from diagrams.aws.database import Neptune
from diagrams.openstack.storage import Swift
from diagrams.generic.blank import Blank
from diagrams.onprem.network import Traefik
from diagrams.gcp.iot import IotCore
from diagrams.gcp.storage import GCS
from diagrams.aws.storage import S3
from diagrams.aws.compute import EC2
from diagrams.onprem.client import Users
from diagrams.programming.language import Javascript
from diagrams.custom import Custom
from diagrams.aws.analytics import ElasticsearchService
from diagrams.onprem.search import Solr


with Diagram("Gleaner Activity Diagram", direction="LR", show=False) as med_diag:
 

    with Cluster("Implementation (Scheduling)"):
        af = Custom("Scheduler)", "./content/schedule.png")
        
            
    with Cluster("User and Community"):
        ui = Custom("User Interface", "./content/chrome-devtools.png")
        with Cluster("Authoring and Guidance"):
            references = Custom("Book", "./content/webdocumentation.png")  
            with Cluster("Validation"):
                sh = Custom("SHACL Notebook", "./content/jupyter.png")
                reports = Custom("Publisher reports", "./content/document.png") 
                sh >> reports 


    with Cluster("Implementation (Data)"):
        oss = S3("Object Store")
        with Cluster("Docker orchestration"):
            ts = Custom("Graph", "./content/jena.png")
            es = Solr("Text index")
            nabu = Go("Nabu")

    with Cluster("Implementation (Indexer)"):
        with Cluster("Docker orchestration"):
            gleaner = Go("Gleaner")
            headless = Custom("Headless Chrome", "./content/chrome-devtools.png")



        
    # Airflow
    af >> gleaner
    af >> nabu
                
    # Indexing
    # rb >> sm >> resources >>  Edge(label="indexing") << gleaner
    gleaner >> oss 
    gleaner >>  Edge(color="darkgreen",  style="dashed") <<  headless
    
    # Validation and reporting
    sh >> oss

    # Object store to Triplestore syncing (inclujding pruning)
    oss >> nabu >> ts
    nabu >> es
    
    # Object store hosed UI
    oss >> ui
    ui >> es
    ui >> ts
    # ui >> references

 
    
    
med_diag
````
